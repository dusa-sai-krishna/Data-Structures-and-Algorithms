```python
import queue
class GenericTreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
def takeInputGT():
    print("Enter root data:")
    rootData=int(input())
    if rootData==-1:
        return None
    root=GenericTreeNode(rootData)
    print("No of children for {}:".format(rootData))
    numChild=int(input())
    
    for i in range(numChild):
        child=takeInputGT()
        root.children.append(child)
    return root

def printGT(root):
    if root is None:
        return None
    #print root Data
 
    print("{}:".format(root.data),end="")
    #print children Data
    for child in root.children:
        print("{},".format(child.data),end="")
    print()
    #calling its children
    for child in root.children:
        printGT(child)
        
def levelWiseInputGT():
    q=queue.Queue()
    rootData=int(input("Enter the root data:"))
    if rootData==-1:
        return None
    root=GenericTreeNode(rootData)
    q.put(root)
    while (not(q.empty())):
        currentNode=q.get()
        print("No of children for {}:".format(currentNode.data))
        numChild=int(input())
        if numChild!=0:
            for i in range(numChild):
                print("Enter the data of {} child".format(i+1))
                childData=int(input())
                child=GenericTreeNode(childData)
                currentNode.children.append(child)
                q.put(child)
    return root
```

```python
def nextLargest(tree, n):
    if tree is None or n<=0:
        return None
    numChild=len(tree.children)
    if numChild==0:
        if tree.data>n:
            return tree.data
        else:
            return None
    if tree.data>n:
        lst=[tree.data]
    else:
        lst=[]
    for child in tree.children:
        ans=nextLargest(child,n)
        if ans!=None:
            lst.append(ans)
    if len(lst)==0:
        return None
    else:
        return min(lst)
```

```python
root=levelWiseInputGT()
printGT(root)
print("\n")
print(nextLargest(root,3))
```
