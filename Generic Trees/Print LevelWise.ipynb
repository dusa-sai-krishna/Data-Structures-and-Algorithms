```python
import queue
class GenericTreeNode:
    def __init__(self,data):
        self.data=data
        self.children=list()
        
def takeInputGT():
    print("Enter root data:")
    rootData=int(input())
    if rootData==-1:
        return None
    root=GenericTreeNode(rootData)
    print("No of children for {}:".format(rootData))
    numChild=int(input())
    
    for i in range(numChild):
        child=takeInputGT()
        root.children.append(child)
    return root

def printGT(root):
    if root is None:
        return None
    #print root Data
 
    print("{}:".format(root.data),end="")
    #print children Data
    for child in root.children:
        print("{},".format(child.data),end="")
    print()
    #calling its children
    for child in root.children:
        printGT(child)
        
def levelWiseInputGT():
    q=queue.Queue()
    rootData=int(input("Enter the root data:"))
    if rootData==-1:
        return None
    root=GenericTreeNode(rootData)
    q.put(root)
    while (not(q.empty())):
        currentNode=q.get()
        print("No of children for {}:".format(currentNode.data))
        numChild=int(input())
        if numChild!=0:
            for i in range(numChild):
                print("Enter the data of {} child".format(i+1))
                childData=int(input())
                child=GenericTreeNode(childData)
                currentNode.children.append(child)
                q.put(child)
    return root
```

```python
def printLevelWiseTree(tree):
    q=queue.Queue()
    if tree is None:
        return None
    q.put(tree)
    q.put(-1)
    #Traverse the tree and print it
    while(not(q.empty())):
        currentNode=q.get()
        if currentNode==-1:
            if (q.empty()):
                return None
            else:
                q.put(-1)
        else:
            print()
            print("{}:".format(currentNode.data),end="")
            numChild=len(currentNode.children)
            if numChild>0:
                for i in range(numChild-1):
                    print("{},".format(currentNode.children[i].data),end="")
                    q.put(currentNode.children[i])
                print("{}".format(currentNode.children[-1].data), end="")
                q.put(currentNode.children[-1])
```

```python
root=levelWiseInputGT()
printGT(root)
```

```python
printLevelWiseTree(root)
```
