```python

from sys import stdin

#Following is the Node class already written for the Linked List
class Node :
    def __init__(self, data) :
        self.data = data
        self.next = None

def length(head):
    temp=head
    i=0
    while temp is not None:
        i+=1
        temp=temp.next
    return i


def appendLastNToFirst(head,tail, n) :
    #Your code goes here
   len=length(head)
   if n>=len:
       return head
   if n==0:
       return head
   n=len-n
   prev=None
   curr=head
   for j in range(n):
       prev=curr
       curr=curr.next
   prev.next=None
   tail.next=head
   return curr
    

























#Taking Input Using Fast I/O
def takeInput() :
    head = None
    tail = None

    datas = [int(i) for i in input().split()]

    i = 0
    while (i < len(datas)) and (datas[i] != -1) :
        data = datas[i]
        newNode = Node(data)

        if head is None :
            head = newNode
            tail = newNode

        else :
            tail.next = newNode
            tail = newNode

        i += 1
    arr=[head,tail]
    return arr


#to print the linked list 
def printLinkedList(head) :

    while head is not None :
        print(head.data, end = " ")
        head = head.next

    print()


#main
t = int(input())

while t > 0 :

    arr = takeInput()
    head=arr[0]
    tail=arr[1]
    n = int(input())

    head = appendLastNToFirst(head,tail, n)
    printLinkedList(head)

    t -= 1 
```
