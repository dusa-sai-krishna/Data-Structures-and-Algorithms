```python
import queue

class BinaryTreeNode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
def printBT(root):
    if root is None:
        return 
    print("{}:".format(root.data),end="")
    if root.left!=None:
        print("L-{},".format(root.left.data),end="")
    if root.right!=None:
        print("R-{}".format(root.right.data),end="")
    print()
    printBT(root.left)
    printBT(root.right)

    
def numNodes(root):
    if root is None:
        return 0
    leftcount=numNodes(root.left)
    rightcount=numNodes(root.right)
    return leftcount+rightcount+1

def takeLevelWiseInput():
    #Create a queue
    q=queue.Queue()
    #Take Root Input
    rootData=int(input("Enter the root data:"))
    #check whether its -1
    if rootData==-1:
        return None
    root=BinaryTreeNode(rootData)
    #place it in queue
    q.put(root)
    #Create a while loop until queue is empty
    while (not(q.empty())):
        currentNode=q.get()
    #Take leftChild Data
        print("Enter the left Child data for the node {}:".format(currentNode.data))
        leftChildData=int(input())
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            currentNode.left=leftChild
            q.put(leftChild)
    #Take RightChild Data
        print("Enter the right Child data for the node {}:".format(currentNode.data))
        rightChildData=int(input())
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            currentNode.right=rightChild
            q.put(rightChild)
    return root  
            
            
def levelWisePrintBT(root):
    #Create a Queue
    q=queue.Queue()
    #Check whether root is not None
    if root is None:
        return None
    #put root into queue
    q.put(root)
    #Create a While loop
    while (not(q.empty())):
        currentNode=q.get()
        print("{}:".format(currentNode.data),end="")
    #print left child
        leftChild=currentNode.left
        if leftChild is not None:
            q.put(leftChild)
            ld=leftChild.data
            print("L{},".format(ld),end="")
    #print right child
        rightChild=currentNode.right
        if rightChild is not None:
            q.put(rightChild)
            rd=rightChild.data
            print("R{}".format(rd),end="")
        print()
```

```python
def level(root, ptr, lev):

    # Base Case
    if root is None:
        return 0
    if root.data == ptr:
        return lev

    # Return level if Node is present in left subtree
    l = level(root.left, ptr, lev+1)
    if l != 0:
        return l

    # Else search in right subtree
    return level(root.right, ptr, lev+1)

def isSiblings(root,p,q):
    if root is None:
        return False
    if root.left is not None and root.right is not None:
        if root.left.data==p and root.right.data==q:
            return True
        if root.left.data == q and root.right.data == p:
            return True
    left=isSiblings(root.left,p,q)
    right=isSiblings(root.right,p,q)
    if left or right:
        return True
    return False


def checkCousins(root,p,q):
    if root is None:
        return False
    #height check
    p_h=level(root,p,1)
    q_h=level(root,q,1)
    if p_h==0 or q_h==0:
        return False
    if p_h!=q_h:
        return False
    return not(isSiblings(root,p,q))
```
