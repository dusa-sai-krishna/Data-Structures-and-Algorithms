```python
import queue

class BinaryTreeNode:
    
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
def printBT(root):
    if root is None:
        return 
    print("{}:".format(root.data),end="")
    if root.left!=None:
        print("L-{},".format(root.left.data),end="")
    if root.right!=None:
        print("R-{}".format(root.right.data),end="")
    print()
    printBT(root.left)
    printBT(root.right)

    
def numNodes(root):
    if root is None:
        return 0
    leftcount=numNodes(root.left)
    rightcount=numNodes(root.right)
    return leftcount+rightcount+1

def takeLevelWiseInput():
    #Create a queue
    q=queue.Queue()
    #Take Root Input
    rootData=int(input("Enter the root data:"))
    #check whether its -1
    if rootData==-1:
        return None
    root=BinaryTreeNode(rootData)
    #place it in queue
    q.put(root)
    #Create a while loop until queue is empty
    while (not(q.empty())):
        currentNode=q.get()
    #Take leftChild Data
        print("Enter the left Child data for the node {}:".format(currentNode.data))
        leftChildData=int(input())
        if leftChildData!=-1:
            leftChild=BinaryTreeNode(leftChildData)
            currentNode.left=leftChild
            q.put(leftChild)
    #Take RightChild Data
        print("Enter the right Child data for the node {}:".format(currentNode.data))
        rightChildData=int(input())
        if rightChildData!=-1:
            rightChild=BinaryTreeNode(rightChildData)
            currentNode.right=rightChild
            q.put(rightChild)
    return root  
            
            
def levelWisePrintBT(root):
    #Create a Queue
    q=queue.Queue()
    #Check whether root is not None
    if root is None:
        return None
    #put root into queue
    q.put(root)
    #Create a While loop
    while (not(q.empty())):
        currentNode=q.get()
        print("{}:".format(currentNode.data),end="")
    #print left child
        leftChild=currentNode.left
        if leftChild is not None:
            q.put(leftChild)
            ld=leftChild.data
            print("L{},".format(ld),end="")
    #print right child
        rightChild=currentNode.right
        if rightChild is not None:
            q.put(rightChild)
            rd=rightChild.data
            print("R{}".format(rd),end="")
        print()
```
