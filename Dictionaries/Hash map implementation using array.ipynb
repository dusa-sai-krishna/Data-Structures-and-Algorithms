```python
#hash map implementation
class MapNode:
    def __init__(self,key,value):
        self.key=key
        self.value=value
        self.next=None



class Map:
    def __init__(self):
        self.bucket_size=10
        self.bucket=[None for i in range(self.bucket_size)]
        self.count=0
        
    def getHashIndex(self,key):
        return (abs(hc(key)%self.bucket_size))
    
    def rehash(self):
        temp=self.bucket
        self.bucket=[None for i in range(2*self.bucket_size)]
        self.bucket_size=2*self.bucket_size
        head=temp[0]
        for head in temp:
            while head is not None:
                self.insert(head.key,head.value)
                head=head.next
       
    def insert(self,key,value):
        hc=hash(key)
        index=abs(hc%self.bucket_size)
        head=self.bucket[index]
        curr=head
        while curr is not None:
            if curr==key:
                curr.value=value    
                return
            curr=curr.next
        newNode=MapNode(key,value)
        newNode.next=head
        self.bucket[index]=newNode
        self.count+=1
        loadFactor=self.count/self.bucket_size
        if loadFactor>=0.7:
            self.rehash()
    def loadFactor(self):
        return (self.count/self.bucket_size)
    
    def remove(self,key):
        #index generation
        hc=hash(key)
        index=abs(hc%self.bucket_size)
        #removing it
        head=self.bucket[index]
        curr=head
        prev=None
        while curr is not None:
            if curr.key==key:
                next=curr.next
                if prev==None:
                    curr=curr.next
                    self.bucket[index]=curr
                else:
                    prev.next=head.next
                self.count-=1
                return head.value
               
            else:
                prev=curr
                curr=curr.next
        return None
        
    def getValue(self,key):
        #index generation
        hc=hash(key)
        index=abs(hc%self.bucket_size)
        #removing it
        head=self.bucket[index]
        curr=head
        while curr is not None:
            if curr.key==key:
                return curr.value
            curr=curr.next
        return None
        
    def search(self,key):
        #index generation
        hc=hash(key)
        index=abs(hc%self.bucket_size)
        #removing it
        head=self.bucket[index]
        curr=head
        while curr is not None:
            if curr.key==key:
                return True
        return False

    
    def size(self):
        return self.count
```
