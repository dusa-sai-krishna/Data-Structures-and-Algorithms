```python
class BinaryTreeNode:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None
    
    
```

```python
class BST:
    def __init__(self):
        self.root=None
        self.numNodes=0
     
    def insertHelper(self,root,data):
        if root is None:
            root=BinaryTreeNode(data)
            return root
        if root.data>data:
            #Call left
            root.left=self.insertHelper(root.left,data)
            return root
        if root.data<=data:
            #Call right
            root.right=self.insertHelper(root.right,data)
            return root
    
    def insert(self,data):
        self.numNodes+=1
        self.root=self.insertHelper(self.root,data)
    
    def minNode(self,root):
        if root is None:
            return 10000
        if root.left==None:
            return root.data
        return self.minNode(root.left)
    def deleteHelper(self,root,data):
        if root is None:
            return False,None
        if root.data>data:
            #call left
            deleted,newleftTree=self.deleteHelper(root.left,data)
            root.left=newleftTree
            return deleted,root
        if root.data<data:
            deleted,newrightTree=self.deleteHelper(root.right,data)
            root.right=newrightTree
            return deleted,root
        if root.data==data:
            #Leaf Node
            if root.left is None and root.right is None:
                return True,None
            #One child
            if root.left is not None and root.right is None:
                return True,root.left
            if root.left is None and root.right is not None:
                return True,root.right
            #Two childs
            rightMin=self.minNode(root.right)
            root.data=rightMin
            deleted,newRightTree=self.deleteHelper(root.right,rightMin)
            root.right=newRightTree
            return deleted,root
    def deleteData(self,data):
        ans,self.root=self.deleteHelper(self.root,data)
        if ans:
            self.numNodes-=1
        return ans
    
    def printTreeHelper(self,root):
        if root is None:
            return None
        print("{}:".format(root.data),end="")
        if root.left is not None:
            print("L{},".format(root.left.data),end="")
        if root.right is not None:
            print("R{},".format(root.right.data),end="")
        print()
        self.printTreeHelper(root.left)
        self.printTreeHelper(root.right)
    def printTree(self):
        return self.printTreeHelper(self.root)
    
    def isPresentHelper(self,root,data):
        if root is None:
            return None
        if root.data==data:
            return True
        if root.data>data:
            #Look left
            return self.isPresentHelper(root.left,data)
        else:
            #Look right
            return self.isPresentHelper(root.right,data)
        
    def isPresent(self,data):
        return self.isPresentHelper(self.root,data)
    
    def count(self):
        return self.numNodes
```

```python
bst=BST()
lst=[10,5,7,6,8,12,11,15]
for i in lst:
	bst.insert(i)
bst.printTree()
print(bst.count())
print(bst.deleteData(8))
bst.printTree()
print(bst.deleteData(5))
bst.printTree()
```
